```python
import cv2
import matplotlib.pyplot as plt
import pyautogui as pag
import time
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model  = 'frozen_inference_graph.pb'
print('\n'.join(f'{m.__name__}=={m.__version__}' for m in globals().values() if getattr(m, '__version__', None)))
model  = cv2.dnn_DetectionModel(frozen_model,config_file)
classLabels  =[]
file_name = 'Labels.txt'
with open(file_name,'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')

model.setInputSize(320,320)
model.setInputScale(1.0/127.5)
model.setInputMean(127.5)
model.setInputSwapRB(True)
img = cv2.imread('test1.png')
ClassIndex, confidence, bbox  =  model.detect(img,confThreshold = 0.5)
font_scale =3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(),bbox):
    cv2.rectangle(img,boxes,(255,0,0),2)
    cv2.putText(img,classLabels[ClassInd-1],(boxes[0] +10,boxes[1]+40), font, fontScale =font_scale,color=(0,255,0), thickness =3)


cap = cv2.VideoCapture("industry.mp4")
#check is opened correctly
if not cap.isOpened():
     cap = cv2.VideoCapture(0)
    
if not cap.isOpened():
     raise IOError("cannot open video")
    
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
print('ankur')
while True:
    ret,frame  = cap.read()
    start_point = (5, 100)
    end_point = (500, 1000)
    color = (255, 0, 0)
    thickness = 2
    cv2.rectangle(frame, start_point, end_point, color, thickness)
    cv2.putText(frame, classLabels[ClassInd-2],(100,100), font, fontScale=font_scale, color=(0,255,0), thickness =1)
    ClassIndex, confidence, bbox =  model.detect(frame,confThreshold =0.55)
    Count = 0
    print(ClassIndex)
    if(len(ClassIndex)!=0):
        for classInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if(ClassInd<=80):
                    cv2.rectangle(frame,boxes,(255,0,0), 2)
                    print(time.perf_counter())
                    cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+20,boxes[1]+5), font, fontScale=font_scale, color=(0,255,0), thickness =1)
                    
    cv2.imshow('Object detection tutorial', frame)
    if cv2.waitKey(2) & 0xFF ==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
